```python
import matplotlib.pyplot as plt
import numpy as np
```

# Scatter Plot

```python
np.random.seed(42)
x = np.random.rand(500) * 10
y = np.random.rand(500) * 100
sizes = np.random.rand(50) * 150

plt.scatter(x, y, c="red", s=100, alpha=0.6, edgecolor="black")
plt.show()
```

# Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, c="red", linestyle="--", linewidth=3, marker="s", markersize=10)
plt.show()
```

# Bar Plot

```python
categories = ["Apples", "Banans", "Cherries", "Dates"]
values = [23, 45, 56, 78]

plt.bar(categories, values, color="red", width=0.5, align="edge", edgecolor="black")
plt.show()
```

# Histogram

```python
np.random.seed(42)
data = np.random.randn(1000)

plt.hist(data, bins=30, color="red", edgecolor='black', alpha=0.7, histtype='stepfilled')
plt.show()
```

# Pie Chart

```python
labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [90, 15, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = [0, 0.1, 0, 0.2]

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=140, explode=explode, shadow=True)
plt.show()
```

## Titles and Labels

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave Example", fontsize=18, fontweight="bold", color="darkred")
plt.xlabel("X-axis (radians)", fontsize=14)
plt.ylabel("sin(x)", fontsize=14)
plt.show()
```

## Ticks and Tick Labels

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.plot(x, y)

plt.xticks(np.arange(0, 11, 1), fontsize=10, rotation=45)
plt.yticks(np.arange(-1, 1.1, 0.5), fontsize=10)

plt.title("Cosine Wave Example with Custom Ticks", fontsize=18, fontweight="bold", color="darkred")
plt.xlabel("X-axis (radians)", fontsize=14)
plt.ylabel("cos(x)", fontsize=14)

plt.grid(True)
plt.show()
```

## Colors and Markers

```python
x = np.linspace(0, 10, 100)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1, color="magenta", marker="s")
plt.plot(x, y2, color="green", linestyle="--", marker="o")

plt.xticks(np.arange(0, 11, 1), fontsize=10, rotation=45)
plt.yticks(np.arange(-1, 1.1, 0.5), fontsize=10)

plt.title("Cosine and Sin Wave Example with Custom Colors and Markers", fontsize=18, fontweight="bold", color="darkred")
plt.xlabel("X-axis (radians)", fontsize=14)
plt.ylabel("Function Value", fontsize=14)
plt.grid(True)
plt.show()
```

## Figure and Axes

```python
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111)

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y)

ax.set_title("Sine Wave in a Single Axes")
ax.set_xlabel("X-axis (radians)")
ax.set_ylabel("sin(x)")
ax.grid(True)

plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 10, 100)
y1 = np.sqrt(x)
y2 = np.sin(x)
y3 = np.cos(x)
y4 = np.tan(x)

axes[0, 0].plot(x, y1)
axes[0, 1].plot(x, y2)
axes[1, 0].plot(x, y3)
axes[1, 1].plot(x, y4)
axes[1, 1].set_ylim(-10, 10)

axes[0, 0].set_xlabel("X-axis")
axes[0, 1].set_xlabel("X-axis")
axes[1, 0].set_xlabel("X-axis")
axes[1, 1].set_xlabel("X-axis")

axes[0, 0].set_ylabel("Y-axis")
axes[0, 1].set_ylabel("Y-axis")
axes[1, 0].set_ylabel("Y-axis")
axes[1, 1].set_ylabel("Y-axis")

axes[0, 0].set_title("Square_Root(x)")
axes[0, 1].set_title("Sin(x)")
axes[1, 0].set_title("Cos(x)")
axes[1, 1].set_title("Tan(x)")

plt.tight_layout()
plt.show()
```

## Legend

```python
x = np.linspace(0, 10, 100)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1, color="magenta", marker="s", label="Cos(x)")
plt.plot(x, y2, color="green", linestyle="--", marker="o", label="Sin(x)")

plt.xticks(np.arange(0, 11, 1), fontsize=10, rotation=45)
plt.yticks(np.arange(-1, 1.1, 0.5), fontsize=10)

plt.title("Cosine and Sin Wave Example with Custom Colors and Markers", fontsize=18, fontweight="bold", color="darkred")
plt.xlabel("X-axis (radians)", fontsize=14)
plt.ylabel("Function Value", fontsize=14)

plt.legend(loc="best", fontsize=10, title="Functions")

plt.grid(True)
plt.show()
```
